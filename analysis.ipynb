```python
#Changing the output format of numbers
def format(number):
    if number >= 1_000_000_000:
        formatted_number = f"{number / 1_000_000_000:.2f}B"
    elif number >= 1_000_000:
        formatted_number = f"{number / 1_000_000:.2f}M"
    elif number >= 1_000:
        formatted_number = f"{number / 1_000:.2f}K"
    else:
        formatted_number = f"{number:.2f}"

    # Add comma separation
    parts = formatted_number.split('.')
    integer_part = parts[0]
    decimal_part = parts[1] if len(parts) > 1 else ''
    
    if 'B' in integer_part:
        integer_part = integer_part.replace('B', '')
        suffix = 'B'
    elif 'M' in integer_part:
        integer_part = integer_part.replace('M', '')
        suffix = 'M'
    elif 'K' in integer_part:
        integer_part = integer_part.replace('K', '')
        suffix = 'K'
    else:
        suffix = ''
    
    integer_part = f"{int(integer_part):,}"

    return f"{integer_part}.{decimal_part}{suffix}"
```

```python
#Checking which is the most held security EUR
grouped_df_eur = port_eur.groupby('Security Description')['Market Value'].sum().reset_index()
# Create a pie chart

fig = px.pie(grouped_df_eur, names='Security Description', values='Market Value', title='Sum of Market Value by Security')
fig.update_traces(textposition='inside')
fig.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')

# Show the plot
fig.show()
```

```python
#Checking which is the most held security USD
grouped_df_usd = port_usd.groupby('Security Description')['Market Value'].sum().reset_index()
# Create a pie chart

fig2 = px.pie(grouped_df_usd, names='Security Description', values='Market Value', title='Sum of Market Value by Security')
fig2.update_traces(textposition='inside')
fig2.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')

# Show the plot
fig2.show()
```

```python
mean_value_usd = port_usd['Coupon'].mean()
std_dev_usd = port_usd['Coupon'].std()
print(f"Mean: {mean_value_usd}")
print(f"Standard Deviation: {std_dev_usd}")
```

```python
# Mean value of coupon rates and their standard deviation

fig3 = go.Figure()

fig3.add_trace(go.Bar(
    x=['Mean'],
    y=[mean_value_usd],
    name='Mean',
    marker_color='blue'
))

# Add standard deviation
fig3.add_trace(go.Bar(
    x=['Standard Deviation'],
    y=[std_dev_usd],
    name='Standard Deviation',
    marker_color='red'
))

# Update layout
fig3.update_layout(
    title='Mean and Standard Deviation of Coupon Values in USD',
    xaxis_title='Statistics',
    yaxis_title='Value',
    showlegend=True
)

# Show the plot
fig3.show()
```

```python
mean_value_eur = port_eur['Coupon'].mean()
std_dev_eur = port_usd['Coupon'].std()
print(f"Mean: {mean_value_eur}")
print(f"Standard Deviation: {std_dev_eur}")
```

```python
# Mean value of coupon rates and their standard deviation

fig4 = go.Figure()

fig4.add_trace(go.Bar(
    x=['Mean'],
    y=[mean_value_eur],
    name='Mean',
    marker_color='blue'
))

# Add standard deviation
fig4.add_trace(go.Bar(
    x=['Standard Deviation'],
    y=[std_dev_eur],
    name='Standard Deviation',
    marker_color='red'
))

# Update layout
fig4.update_layout(
    title='Mean and Standard Deviation of Coupon Values in USD',
    xaxis_title='Statistics',
    yaxis_title='Value',
    showlegend=True
)

# Show the plot
fig4.show()
```
